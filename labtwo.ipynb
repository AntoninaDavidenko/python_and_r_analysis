```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(42)
num_shops = np.random.randint(20, 100, 50)

sales = 50000 * num_shops + np.random.normal(0, 500000, 50)  # Додаємо шум

df = pd.DataFrame({'Магазини': num_shops, 'Продажі': sales})

X = df[['Магазини']]
y = df['Продажі']

model = LinearRegression()
model.fit(X, y)

# Прогноз для 73 магазинів
shops_in_area = pd.DataFrame([[73]], columns=['Магазини'])
predicted_sales = model.predict(shops_in_area)

print(f"Прогнозований річний обсяг продажів для 73 магазинів: {predicted_sales[0]:,.2f} грн")

# Візуалізація
plt.scatter(df['Магазини'], df['Продажі'], label="Реальні дані", color="blue")
plt.plot(df['Магазини'], model.predict(X), color="red", label="Лінія регресії")
plt.scatter(73, predicted_sales, color="green", label="Прогноз (73 магазини)", marker="X", s=150)
plt.xlabel("Кількість магазинів")
plt.ylabel("Річний чистий обсяг продажів (грн)")
plt.title("Лінійна регресія для прогнозування продажів")
plt.legend()
plt.grid()
plt.show()
```
